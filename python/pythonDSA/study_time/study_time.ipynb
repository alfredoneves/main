```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
from numpy import genfromtxt

# to avoid mess in jupyter when using matplotlib
%matplotlib inline 

# read csv file and display basic information
file0 = pd.read_csv("control.csv", sep=";")
print(type(file0))
print(file0.shape) # lines and columns (without the column names)

# converts the columns is lists to use in matplotlib
months = list(file0["Months"])
hours = list(file0["Hours"])

# plotting 
plt.xlabel('Months of the year')
plt.ylabel('Quantity of hours I studied')
plt.title('Graphic to represent my study in 2022')

plt.style.use('ggplot')
plt.plot(months, hours, color="green")
```
